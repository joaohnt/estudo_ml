```python
from sklearn.neural_network import MLPClassifier
import pickle
from sklearn.metrics import accuracy_score, classification_report
```

# base credit data - 99,8% ACC

```python
with open('credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
 rede_neural_credit = MLPClassifier(max_iter=1500, verbose=True, tol=0.0000100,
                                    solver='adam', activation='relu', hidden_layer_sizes=(2,2))
 rede_neural_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = rede_neural_credit.predict(X_credit_teste)
```

```python
previsoes
```

```python
y_credit_teste
```

```python
accuracy_score(y_credit_teste, previsoes)
```

```python
 print(classification_report(y_credit_teste, previsoes))
```

# base census - 81.6% ACC

```python
with open('census.pkl', 'rb') as f:
  X_census_treinamento, y_census_treinamento, X_census_teste, y_census_teste = pickle.load(f)
```

```python
# (108 + 1) / 2 == 54.5 == 55
rede_neural_census = MLPClassifier(verbose=True, max_iter = 1000, tol=0.000010, hidden_layer_sizes=(55,55))
rede_neural_census.fit(X_census_treinamento, y_census_treinamento)
```

```python
previsoes = rede_neural_census.predict(X_census_teste)
previsoes
```

```python
y_census_teste
```

```python
accuracy_score(y_census_teste, previsoes)
```

```python
print(classification_report(y_census_teste, previsoes))
```

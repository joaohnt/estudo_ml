```python
import pandas as pd
from apyori import apriori
```

# base mercado 1

```python
base_mercado1 = pd.read_csv('mercado.csv', header=None)
base_mercado1
```

```python
transacoes = []
for i in range(len(base_mercado1)):
    linha = [str(base_mercado1.values[i, j]) for j in range(base_mercado1.shape[1]) if str(base_mercado1.values[i, j]) != 'nan']    # removendo os NaN da lista
    transacoes.append(linha)
```

```python
transacoes
```

```python
regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.8, min_lift = 2)
resultados = list(regras)
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
  s = resultado[1]
  result_rules = resultado[2]
  for result_rule in result_rules:
    a = list(result_rule[0]) # se
    b = list(result_rule[1]) # entao
    print(a, '-->', b)
    c = result_rule[2]
    l = result_rule[3]
    print('support: ', s)
    print('confidence: ', c)
    print('lift: ', l)
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
```

```python
rules_df.sort_values(by = 'lift', ascending = False)
```

```python
transacoes = []
for i in range(len(base_mercado2)):
    linha = [str(base_mercado2.values[i, j]) for j in range(base_mercado2.shape[1]) if str(base_mercado2.values[i, j]) != 'nan']    # removendo os NaN da lista
    transacoes.append(linha)
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
  s = resultado[1]
  result_rules = resultado[2]
  for result_rule in result_rules:
    a = list(result_rule[0]) # se
    b = list(result_rule[1]) # entao
    print(a, '-->', b)
    c = result_rule[2]
    l = result_rule[3]
    print('support: ', s)
    print('confidence: ', c)
    print('lift: ', l)
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```
